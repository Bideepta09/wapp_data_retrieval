```python
#Problem 1(List of unique members who has interacted at least once in this group)
fhand = open("chat.txt",encoding = 'utf-8')

names = []
for line in fhand:
    start = line.find("-")
    start1 = line.find(":",line.find(":") + 1)
    if start > 0 and start1 > 0 and ": " and "/2" in line:
        name = line[start + 2: start1]
        if name not in names:
            names.append(name)
print(names)
```

```python
#Problem 2(Extracting entire conversation which happened on a particular date)
def conversation_by_date(file,date):
    fhand = open(file,encoding = 'utf-8')
    for line in fhand:
        st = line.find(",")
        if st > 0 and "/2" in line:
            d = line[:st]
            if date == d:
                line = line.rstrip()
                print(line)
```

```python
conversation_by_date("chat.txt","18/02/2022")
```

```python
#Problem 3(Extracting all the posts done by a particular person)
def conversation_by_person(file,person):
    fhand = open(file,encoding = 'utf-8')
    for line in fhand:
        start = line.find("-")
        start1 = line.find(":",line.find(":") + 1)
        if start > 0 and start1 > 0 and ": " and "/2" in line:
            name = line[start + 2: start1]
            if person == name:
                line = line.rstrip()
                print(line)
```

```python
conversation_by_person("chat.txt","Gourab Nath")
```

```python
#Problem 4(Which part of the day does this group experience most conversation)
fhand = open("chat.txt",encoding = 'utf-8')
am = 0
pm = 0
for line in fhand:
    pos = line.find(", ")
    if pos > 0 and line[pos + 3] == ":":
        t = line[pos + 1 : pos + 3]
        if int(t) > 12:
            pm += 1
        else:
            am += 1
if am > pm:
    print("The group is more active in Day(AM hours)")
else:
    print("The group is more active in Night(PM hours)")
```

```python
#Problem 5(Frequency Distribution of the hours)
fhand = open("chat.txt",encoding = 'utf-8')
hours = []
count = {}
for line in fhand:
    s = line.find(", ")
    if s > 0 and "/2" in line:
        hour = line[s + 2 : s + 4]
        hours.append(hour)
for hour in hours:
    if hour not in count:
        count[hour] = 1
    else:
        count[hour] = count[hour] + 1
print("The frequency distribution of hours:\n")
print(count)
```

```python
#Problem 6(Average length of the messages posted by a particular person)
def avg_length_by_person(file,person):
    count = 0
    avg_length = 0
    fhand = open("chat.txt",encoding = 'utf-8')
    for line in fhand:
        start = line.find("-")
        start1 = line.find(":",line.find(":") + 1)
        if start > 0 and start1 > 0 and ": " and "/2" in line:
            name = line[start + 2: start1]
            if person == name:
                count += 1
                avg_length = avg_length + len(line) - start1 - 2
    print("Average length of the messages by the person is: " ,avg_length/count)
```
